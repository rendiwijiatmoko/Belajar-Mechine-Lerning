```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
y = np.sqrt(x)
y1 = np.sin(x)
y2 = np.cos(x)
```

## Inisialisasi Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x,y1)
ax.plot(x,y2)
```

## shortcut

```python
plt.figure(figsize=(6,6))
plt.plot(x,y1, 'r:')
plt.plot(x,y2, 'bo', linewidth=8)

plt.xlim(0,10)
plt.ylim(-2,2)
plt.axis('equal')
plt.title('sin dan cos')
plt.xlabel("sumbu x")
plt.ylabel("sumbu y");
```

## Scatter Plot

## Subplot

### Matlab Style

```python
plt.figure(figsize=(15,10))

plt.subplot(231) # kalau ada dua digit pisahkan pakai koma aja contoh (3,5,10)
plt.hist(numbers, bins=80, density=1);

plt.subplot(233)
plt.hist(numbers, bins=80, density=1);

plt.subplot(235)
plt.scatter(x,y, c='r', s=10)
plt.axis('equal');

plt.subplot(236)
plt.hist(numbers, bins=80, density=1, cumulative=True);
```

### Matplotlib Style

```python
fig, ax =plt.subplots(2,3, figsize=(15,10))

ax[0,0].hist(numbers, bins=80, density=1);

ax[0,1].axis('off')

ax[0,2].hist(numbers, bins=80, density=1);

ax[1,0].axis('off')

ax[1,1].scatter(x,y, c='r', s=10)
ax[1,1].axis('equal');


ax[1,2].hist(numbers, bins=80, density=1, cumulative=True);
```

```python
fig, ax =plt.subplots(2,3, figsize=(15,10), sharey=True, sharex=True)

ax[0,0].hist(numbers, bins=80, density=1);

ax[0,2].hist(numbers, bins=80, density=1);

ax[1,1].scatter(x,y, c='r', s=10)

ax[1,2].hist(numbers, bins=80, density=1, cumulative=True);
```

## Text dan Anotation

```python
numbers = np.random.randn(10000)
```

```python
plt.figure(figsize=(6,6))
plt.hist(numbers, bins=80, density=1);
plt.ylim(0,1)
plt.axvline(0, color='r');
plt.text(0,0.8,"mean=0", ha='center')
plt.annotate("mean=0", xy=(0,0.8), xytext=(1,0.5), ha="center", )
```
