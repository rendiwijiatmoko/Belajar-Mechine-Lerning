```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/linear.csv')
df.head()
```

```python
df1 = df[df.tipe == "data1"]
df2 = df[df.tipe == "data2"]

plt.scatter(df1.X, df1.y)
plt.scatter(df2.X, df2.y);
```

## Seaborn

```python
plt.figure(figsize=(6,6))
sns.scatterplot("X","y",data=df, hue="tipe");
```

```python
plt.figure(figsize=(6,6))
sns.lmplot("X","y",data=df, hue="tipe");
```

```python
plt.figure(figsize=(6,6))
sns.distplot(df1.X, bins=20);
```

## Seaborn & Pandas DataFrame

### Statistikal Summary

```python
df = pd.read_csv('data/cereal.csv', index_col="name")
df.head()
```

#### Missing Value

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.isna(), cmap='Blues', cbar=False);
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), cmap='bwr', cbar=False, vmin=-1, vmax=1, annot=True, square=True, fmt=".1f");
```

## Pairplot

```python
sns.pairplot(df, vars=["calories", "fat", "rating"], kind='reg');
```

```python
sns.jointplot("potass", "fiber",df, marginal_kws={"bins":25}, kind='reg');
```

## Categorical Data

### Persebaran

```python
sns.catplot("mfr", "rating", data=df);
```

```python
sns.boxenplot("mfr", "rating", data=df);
```

```python
sns.violinplot("mfr", "rating", data=df);
```

### Count

```python
sns.catplot("mfr", data=df, kind="count");
```

```python
sns.countplot("mfr", data=df, hue="shelf");
```
